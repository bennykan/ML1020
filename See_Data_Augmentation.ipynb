```python
from keras.preprocessing.image import ImageDataGenerator
import os


base_dir = '/home/jupyter/Data'
train_dir = os.path.join(base_dir, 'train')

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
    )





```

```python
import cv2
import numpy as np

img = cv2.imread('/home/jupyter/Data/train/c0/img_100026.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res = cv2.resize(img, (150, 150),0, 0,  interpolation = cv2.INTER_NEAREST)
plt.imshow(res)
```

```python
from keras.preprocessing.image import array_to_img, img_to_array, load_img
image = img_to_array(res)
image = image.reshape((1,) + image.shape)
new_image = array_to_img(image[0])
```

```python
count = 9
images_flow = train_datagen.flow(image, batch_size=1)
for i, new_images in enumerate(images_flow):
    new_image = array_to_img(new_images[0])
    plt.imshow(new_image)
    plt.show()
    if i >= count:
        break
        
        
```
